```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data_path = "C:\\Users\\Asus\\OneDrive\\Desktop\\Sem4\\ML\\lab\\ex_1\\data.csv" 
df = pd.read_csv(data_path, encoding='unicode_escape')
```

```python
print("First 5 rows:") 
display(df.head()) 
```

```python
print("\nLast 5 rows:") 
display(df.tail()) 
```

```python
print("\nDataset Info:") 
df.info() 
print("\nStatistical Summary:") 
display(df.describe(include='all'))
```

```python
print("\nMissing Values per Column:") 
print(df.isnull().sum())
```

```python
product_sales = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
print("\nUnique Products:", df['Description'].nunique())
```

```python
plt.figure(figsize=(12,6)) 
sns.barplot(x=product_sales.values, y=product_sales.index, palette="viridis") 
plt.title("Top 10 Products by Quantity Sold") 
plt.xlabel("Total Quantity Sold") 
plt.ylabel("Product") 
plt.show()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) 
sales_over_time = df.groupby(df['InvoiceDate'].dt.date)['Quantity'].sum() 
plt.figure(figsize=(14,6)) 
sales_over_time.plot(kind='line', marker='o') 
plt.title("Sales Trend Over Time") 
plt.xlabel("Date") 
plt.ylabel("Total Quantity Sold") 
plt.grid(True) 
plt.show()
```

```python
print('MIDHUN P 24BAD069')
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data_path = "C:\\Users\\Asus\\OneDrive\\Desktop\\Sem4\\ML\\lab\\ex_1\\Housing.csv"
df = pd.read_csv(data_path)
```

```python
print("First 5 rows:") 
display(df.head())
```

```python
print("\nDataset Info:") 
df.info()
```

```python
print("\nStatistical Summary:") 
display(df.describe())
```

```python
print("\nColumn Names:") 
print(df.columns)
```

```python
print("\nMissing Values per Column:") 
print(df.isnull().sum())
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100 
print("\nPercentage of Missing Values:") 
print(missing_percent)
```

Scatter Plots

```python
plt.figure(figsize=(12,5)) 
sns.scatterplot(x='area', y='price', data=df, alpha=0.5) 
plt.title("Relationship between Living Area (sqft) and Price") 
plt.show() 
plt.figure(figsize=(12,5)) 
sns.scatterplot(x='bedrooms', y='price', data=df, alpha=0.5) 
plt.title("Relationship between Bedrooms and Price") 
plt.show()
```

Heat Map

```python
plt.figure(figsize=(12,8)) 
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f") 
plt.title("Correlation Heatmap of Housing Features") 
plt.show()
```

```python
price_corr = corr_matrix['price'].sort_values(ascending=False) 
print("\nCorrelation of Features with Price:") 
print(price_corr)
```

```python
print('MIDHUN P 24BAD069')
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data_path = r"C:\Users\Asus\OneDrive\Desktop\Sem4\ML\lab\ex_1\marketing_campaign.csv" 
df = pd.read_csv(data_path, sep="\t") # dataset uses tab separator
```

Visualize age Distribution

Histogram

```python
df['Age'] = 2026 - df['Year_Birth'] 
plt.figure(figsize=(12,5)) 
sns.histplot(df['Age'], bins=20, kde=True, color='blue') 
plt.title("Age Distribution of Customers") 
plt.xlabel("Age") 
plt.ylabel("Count") 
plt.show()
```

Box PlOt

```python
plt.figure(figsize=(6,4)) 
sns.boxplot(x=df['Age'], color='orange') 
plt.title("Age Boxplot") 
plt.show()
```

Visualize Income Distribution

Histogram

```python
plt.figure(figsize=(12,5)) 
sns.histplot(df['Income'], bins=20, kde=True, color='green') 
plt.title("Income Distribution of Customers") 
plt.xlabel("Income") 
plt.ylabel("Count") 
plt.show()
```

Boxplot

```python
plt.figure(figsize=(6,4)) 
sns.boxplot(x=df['Income'], color='red') 
plt.title("Income Boxplot") 
plt.show()
```

Visualize Spending Patterns

Bar plot

```python
spending_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'] 
spending = df[spending_cols].sum().sort_values(ascending=False) 
plt.figure(figsize=(10,6)) 
sns.barplot(x=spending.values, y=spending.index, palette="viridis") 
plt.title("Total Spending by Product Category") 
plt.xlabel("Total Amount Spent") 
plt.ylabel("Product Category") 
plt.show()
```

```python
print("\n--- Insights ---") 
print("1. Age distribution shows most customers are between 30–60 years old.") 
print("2. Income distribution reveals a wide range, with some missing values that need imputation.") 
print("3. Spending patterns indicate wines and meat products are the most purchased categories.") 
print("4. Missing values in Income column suggest incomplete account information for some customers.")
```
